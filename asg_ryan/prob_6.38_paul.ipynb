```python
import scipy.stats as stat
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 1000)
ns = 2**np.arange(0,6)

a1,b1 = 1,4
beta_pdf1 = stat.beta.pdf(x, a=a1, b=b1)
mean1 = stat.beta.mean(a=a1, b=b1)
var1 = stat.beta.var(a=a1, b=b1)

a2,b2 = 1,1
beta_pdf2 = stat.beta.pdf(x, a=a2, b=b2)
mean2 = stat.beta.mean(a=a2, b=b2)
var2 = stat.beta.var(a=a2, b=b2)
```

```python
#i. and ii.

#Beta(1,4)
plt.figure(figsize=(15, 5))
plt.plot(x, beta_pdf1, label="Beta(a=%d, b=%d)"%(a1, b1))
for n in ns:
    plt.plot(x, stat.norm.pdf(x, loc=mean1, scale=np.sqrt(var1/n)),
             label="Normal(mean=%.2f, sd=%.3f)"%(mean1, np.sqrt(var1/n)))
plt.legend()
plt.show()

#Beta(1,1)
plt.figure(figsize=(15, 5))
plt.plot(x, beta_pdf2, label="Beta(%d, %d)"%(a2, b2))
for n in ns:
    plt.plot(x, stat.norm.pdf(x, loc=mean2, scale=np.sqrt(var2/n)),
             label="Normal(mean=%.2f, sd=%.3f)"%(mean2, np.sqrt(var2/n)))
plt.legend()
plt.show()
```

```python
# iii. and iv.
for n in ns:
    #Beta(1,4)
    samples1 = stat.beta.rvs(a=a1, b=b1, size=(n, 1000))
    xbar1 = np.mean(samples1, axis=0)
    
    
    #Beta(1,1)
    samples2 = stat.beta.rvs(a=a2, b=b2, size=(n, 1000))
    xbar2 = np.mean(samples2, axis=0)
    
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 4))
    ax[0].hist(xbar1, normed=True, bins=25)
    ax[0].set_title("Beta(%d, %d) with (n: %d)"%(a1, b1, n), size=20)
    ax[1].hist(xbar2, normed=True, bins=25)
    ax[1].set_title("Beta(%d, %d) with (n: %d)"%(a2, b2, n), size=20)
    
    plt.show()
```
